### 对boston数据进行回归

```python
from sklearn import tree
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score

boston = load_boston()
regressor = tree.DecisionTreeRegressor(random_state=0)
cross_val_score(regressor, boston.data, boston.target, cv=10, scoring='neg_mean_squared_error')
```

### 构造带噪声的正弦信号，并进行回归

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree

rng = np.random.RandomState(1)
x = np.sort(5*rng.rand(80, 1), axis=0)
y = np.sin(x).ravel()              # ravel()降维函数
```

```python
regr_1 = tree.DecisionTreeRegressor(max_depth=2)   # 模型1最大深度为2
regr_2 = tree.DecisionTreeRegressor(max_depth=5)   # 模型2最大深度为5
regr_1.fit(x, y)
regr_2.fit(x, y)
```

```python
# np.arrange(开始点，结束点，步长) 生成有序数组的函数
x_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]      # np.newaxis用于增加维度
y_1 = regr_1.predict(x_test)
y_2 = regr_2.predict(x_test)
```

```python
plt.figure()
plt.scatter(x, y, s=20, edgecolors='k', c='darkorange', label='data')
plt.plot(x_test, y_1, color='cornflowerblue', label='max_depth=2', linewidth=2)
plt.plot(x_test, y_2, color='yellowgreen', label='max_depth=5', linewidth=2)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```
